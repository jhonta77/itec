```python
data = {
    'Producto': ['CIFRUT', 'Cargo extra', 'CIFRUT', 'Cargo extra', 'CIFRUT'],
    'Cantidad': [1, 1, 1, 1, 1]
}
df = pd.DataFrame(data)

# Contar la cantidad de productos comprados
producto_counts = df['Producto'].value_counts()

# Calcular los porcentajes
producto_percentages = (producto_counts / producto_counts.sum()) * 100

# Crear la gráfica de barras
fig, ax = plt.subplots()
producto_counts.plot(kind='bar', ax=ax, color='skyblue', alpha=0.8)
ax.set_title('Cantidad de productos comprados')
ax.set_xlabel('Producto')
ax.set_ylabel('Cantidad')
ax.bar_label(ax.containers[0], labels=[f'{p:.1f}%' for p in producto_percentages])

# Mostrar la gráfica en Streamlit
st.pyplot(fig)

# Mostrar la tabla con cantidades y porcentajes
st.write("Tabla de productos comprados:")
st.write(pd.DataFrame({
    'Producto': producto_counts.index,
    'Cantidad': producto_counts.values,
    'Porcentaje': producto_percentages.values
}))

# ...existing code...
```
